```python
import speech_recognition as sr
import pyttsx3
import subprocess
import pyautogui
import datetime
import time
import ctypes
import wikipedia
import webbrowser
import winsound


# Initialize the speech recognition engine
recognizer = sr.Recognizer()

# Initialize the text-to-speech engine
text_to_speech = pyttsx3.init()

# Flag to control whether the program should continue listening
listening = True

                # Functions

def speak(text):
    text_to_speech.say(text)
    text_to_speech.runAndWait()
    

def beep():
    winsound.Beep(1100, 500)  # Frequency: 1000 Hz, Duration: 1000 ms (1 second)

def take_screenshot():   # Screen shot function
    
    # Simulate pressing the Windows key + PrtSc
    pyautogui.hotkey('win', 'prtsc')    
    # pyautogui.screenshot()
    # screenshot.save("screenshot.png")

def welcome():
    
    hour=datetime.datetime.now().hour
    if hour>=0 and hour<12:
        speak("Hello,Good Morning")
        print("Hello,Good Morning...")
    elif hour>=12 and hour<18:
        speak("Hello,Good Afternoon")
        print("Hello,Good Afternoon...")
    else:
        speak("Hello,Good Evening")
        print("Hello,Good Evening...")

# Function to check the Caps Lock state and toggle it based on input
def toggle_caps_lock(desired_state):
    # Get the state of the Caps Lock key using ctypes
    caps_lock_state = ctypes.windll.user32.GetKeyState(0x14)
    
    # Check if the Caps Lock key is in the on state
    if caps_lock_state & 1:
        if desired_state == 0:
            pyautogui.press('capslock')  # Turn Caps Lock off
            print("Caps Lock is now off.")
            speak("Caps Lock is now off.")

        else:
            print("Caps Lock is already on.")
            speak("Caps Lock is already on.")

    else:
        if desired_state == 1:
            pyautogui.press('capslock')  # Turn Caps Lock on
            print("Caps Lock is now on.")
            speak("Caps Lock is now on.")

        else:
            print("Caps Lock is already off.")
            speak("Caps Lock is already off.")


# Function to check the Num Lock state and toggle it based on input
def toggle_num_lock(desired_state):
    # Get the state of the Num Lock key using ctypes
    num_lock_state = ctypes.windll.user32.GetKeyState(0x90)

    # Check if the Num Lock key is in the on state
    if num_lock_state & 1:
        if desired_state == 0:
            pyautogui.press('numlock')  # Turn Num Lock off
            print("Num Lock is now off.")
            speak("Num Lock is now off.")
        else:
            print("Num Lock is already on.")
            speak("Num Lock is already on.")

    else:
        if desired_state == 1:
            pyautogui.press('numlock')  # Turn Num Lock on
            print("Num Lock is now on.")
            speak("Num Lock is now on.")
            
        else:
            print("Num Lock is already off.")
            speak("Num Lock is already off.")


# Function to search Wikipedia and provide a summary
def search_wikipedia(query):
    try:
        # Search Wikipedia for the query
        results = wikipedia.summary(query, sentences=2)
        return results
    except wikipedia.exceptions.DisambiguationError as e:
        # Handle disambiguation pages if needed
        return "There are multiple possible results. Please specify your query."
    except wikipedia.exceptions.PageError as e:
        # Handle cases where the page doesn't exist
        return "I couldn't find any information on that topic."

                                # COMMANDS
def execute_command(command):
    
    global listening  # Use the global listening variable

    if 'hello ghost' in command or 'hello' in command:
            welcome()

    elif any(keyword in command for keyword in ["bye", "stop", "tata"]):
        speak("Goodbye!")
        print("Goodbye!...")
        listening = False  # Set listening to False to stop the loop

    elif 'open youtube' in command:
            webbrowser.open_new_tab("https://www.youtube.com")
            speak("youtube is now open")
            time.sleep(5)

    elif 'open chat gpt' in command:
        webbrowser.open_new_tab("https://chat.openai.com/")
        speak("Chat GPT is now open")
        time.sleep(5)

    elif 'open google drive' in command:
        webbrowser.open_new_tab("https://drive.google.com/drive/my-drive")
        speak("Google drive is now open")
        time.sleep(5)

    elif 'open gmail' in command:
        webbrowser.open_new_tab("https://www.gmail.com")
        speak("Google Mail now open")
        time.sleep(5)

    elif 'time' in command or 'time now' in command:
        timee = datetime.datetime.now().strftime("%I:%M %p")
        print(f"The time is {timee}")
        speak(f"The time is {timee}")

    elif "open" in command or "execute" in command:
        app_name = command.split(" ", 1)[-1]
        try:
            # Attempt to directly open known applications
            subprocess.Popen(app_name)
        except FileNotFoundError:
            # If not found, try searching for the application using Windows key
            pyautogui.hotkey('win', 's')
            time.sleep(2)  # Wait for the search bar to appear
            pyautogui.write(app_name)
            time.sleep(2)  # Wait for search results to appear
            pyautogui.press('enter')

    elif "search" in command: 
        text_to_type = command.split(" ", 1)[-1]
        pyautogui.write(text_to_type)
        pyautogui.press('enter')

    elif "type" in command  or "write" in command:
        text_to_type = command.split(" ", 1)[-1]
        pyautogui.write(text_to_type)

    elif "press" in command:
        key_to_press = command.split(" ", 1)[-1]
        
        if key_to_press == "windows key":
            pyautogui.press('winleft')  # Press the left Windows key
        # or pyautogui.press('winright') to press the right Windows key
        elif key_to_press == "caps lock key":
            pyautogui.press('capslock')
        elif key_to_press == "num lock key":
            pyautogui.press('numlock')
        else:
            pyautogui.press(key_to_press)


    elif "caps lock on" in command:
        toggle_caps_lock(1)  # Toggle Caps Lock based on its current state

    elif "caps lock off" in command:
        toggle_caps_lock(0)  # Toggle Caps Lock based on its current state

    elif "num lock on" in command:
        toggle_num_lock(1)  # Toggle Num Lock based on its current state

    elif "num lock off" in command:
        toggle_num_lock(0)  # Toggle Num Lock based on its current state

    elif 'show desktop' in command:
        # Simulate pressing Win + D to show the desktop or restore windows
        pyautogui.hotkey('win', 'd')
        print("Showed the desktop or restored windows")
        speak("Showed the desktop or restored windows")

    elif "calculate" in command:
        expression = command.split("calculate", 1)[-1]
        result = eval(expression)
        pyautogui.write(str(result))
        pyautogui.press('enter')

    elif "close" in command or "exit" in command:
        # Simulate pressing Alt + F4 to close the active window
        pyautogui.hotkey('alt', 'f4')
        speak("Closing the active window")


    elif "screenshot" in command:
        speak("Taking a screenshot")
        take_screenshot()
        
    elif "copy" in command:
        pyautogui.hotkey('ctrl', 'c')
        speak("Copied")
        
    elif "paste" in command:
        pyautogui.hotkey('ctrl', 'v')
        speak("Pasted")        
    
    elif "select all" in command:
        pyautogui.hotkey('ctrl', 'a')
        speak("All selected")  
    
    elif "new tab" in command:
        # Simulate pressing Ctrl + T to open a new tab in a web browser
        pyautogui.hotkey('ctrl', 't')
        print("Opened a new tab")
        speak("Opened a new tab")
    
    elif "new window" in command:
        # Simulate pressing Ctrl + N to open a new window in a web browser
        pyautogui.hotkey('ctrl', 'n')
        print("Opened a new window")
        speak("Opened a new window")
    
    elif "send" in command:
        pyautogui.press('enter')

    
    elif "incognito window" in command:
        # Simulate pressing Ctrl + Shift + N to open a new incognito window in Google Chrome
        pyautogui.hotkey('ctrl', 'shift', 'n')
        print("Opened a new incognito window")
        speak("Opened a new incognito window")
    
    elif 'next tab' in command:
        # Simulate pressing Ctrl + Tab to move to the next tab in Google Chrome
        pyautogui.hotkey('ctrl', 'tab')
        print("Moved to the next tab")
        speak("Moved to the next tab")

    elif 'refresh' in command:
        pyautogui.rightClick()
        # Delay briefly to allow the context menu to appear (adjust as needed)
        time.sleep(2)
        # Simulate pressing the down arrow key twice
        pyautogui.press('down')
        time.sleep(0.5)  # Adjust the delay between key presses if needed
        pyautogui.press('down')
        time.sleep(0.5)  # Adjust the delay between key presses if needed
        # Simulate pressing Enter
        pyautogui.press('enter')
    
    elif 'right click' in command or 'click right' in command:
        pyautogui.rightClick()
    
    elif 'left click' in command or 'click left' in command:
        pyautogui.click()
    
    elif 'previous tab' in command:
        # Simulate pressing Ctrl + Shift + Tab to move to the previous tab in Google Chrome
        pyautogui.hotkey('ctrl', 'shift', 'tab')
        print("Moved to the previous tab")
        speak("Moved to the previous tab")

    elif 'URL search' in command:
        # Simulate pressing Ctrl + L to focus on the URL search bar in Google Chrome
        pyautogui.hotkey('ctrl', 'l')
        print("Focused on the URL search bar")
        speak("Focused on the URL search bar")

    elif "refresh webpage" in command or "refresh this webpage" in command:
        # Simulate pressing Ctrl + R to refresh the webpage in Google Chrome
        pyautogui.hotkey('ctrl', 'r')
        print("Refreshed")
        speak("Refreshed")

    elif "chrome history" in command:
        # Simulate pressing Ctrl + H to open the browsing history in Google Chrome
        pyautogui.hotkey('ctrl', 'h')
        print("Opened history tab")
        speak("Opened history tab")

    elif "chrome downlaods" in command:
        # Simulate pressing Ctrl + J to open the downloads page in Google Chrome
        pyautogui.hotkey('ctrl', 'j')
        print("Opened downlaods tab")
        speak("Opened downlaods tab")

    elif "wikipedia" in command:
        query = command.replace("wikipedia", "").strip()
        print('Searching Wikipedia...')
        speak('Searching Wikipedia...')
        result = search_wikipedia(query)
        print("According to Wikipedia")
        speak("According to Wikipedia")
        print(result)
        speak(result)
        


    else:
        speak("I'm sorry, no such command.")

# Main loop
while listening:
    with sr.Microphone() as source:
        print("Listening for a command...")
        speak("Ahan")
        
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        command = recognizer.recognize_google(audio).lower()
        print("You said:", command)
        execute_command(command)
        
    except sr.UnknownValueError:
        print("Sorry, could not understand your audio.")
    except sr.RequestError as e:
        print(f"Could not request results; {e}")
```
